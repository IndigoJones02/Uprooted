```python
def clean_std_ltdb(df):
    for i in range(len(df)):
        if df['population'][i] == 0:
            df = df.drop([i])
    df = df.drop(['population', 'non-hisp black', 'non-hisp white', 'native american', 'asian', 'hispanic', 'indian', 'chinese', 'japanese', 'korean', 'vietnamnese', 'mexican', 'puerto rican', 'cuban', 'filipino'], axis=1)
    return df

std_housing_info = clean_std_ltdb(std_ltdb)
# display(std_housing_info)
```

```python
def minority_column(df):
    minority_count, total, immigrant_minority = 0, 0, 0
    m_total, population, w_perc, b_perc, m_perc, i_perc, tracts, w_total = [],[],[],[],[],[],[],[]

    for i in range(len(df)):
        minority_count =  df['non-hisp black'][i]+ df['native american'][i]+ df['hispanic'][i]+ df['asian'][i] + df['puerto rican'][i] + df['cuban'][i]
        
        total =  df['non-hisp white'][i] + df['non-hisp black'][i] + df['native american'][i] + df['hispanic'][i] + df['asian'][i] + df['puerto rican'][i] + df['cuban'][i]  
    
        immigrant_minority =  df['hispanic'][i]+ df['asian'][i] + df['puerto rican'][i] + df['cuban'][i]
       
        if total == 0:
            df = df.drop([i])
        else:
            population.append(total)
            m_total.append(minority_count)
            i_perc.append((immigrant_minority/total)*100)
            m_perc.append((minority_count/total)*100)
            w_perc.append((df['non-hisp white'][i]/total)*100)
            b_perc.append((df['non-hisp black'][i]/total)*100)
            tracts.append(df['tractID'][i])
            w_total.append(df['non-hisp white'][i])
            

        minority_count = 0
        immigrant_minority = 0
        total =0
    return population, m_total, i_perc, b_perc, w_perc, m_perc, df, tracts, w_total

pop_col, min_col, ip_col, bp_col, wp_col, mp_col, df, tracts, w_col = minority_column(std_ltdb)
print()


population_make_up = pd.DataFrame({'tractID': tracts,
                                   'population': pop_col,
                                   'non-hisp white': w_col,
                                   'minorities': min_col,
                                   'non-hisp white %': wp_col,
                                   'non-hisp black %': bp_col,
                                   'immigrant minority %': ip_col,
                                   'minority %': mp_col})

display(population_make_up)
```

```python
path3 = "/Users/hbcuc2-miles/Desktop/Gen-Datasets/LTDB-Stats/LTDB_DP.xlsx"
data3 = pd.read_excel(path3)
# display(data3)

def dtype_conversion(df, col):
    df[col] = df[col].astype(str)
    
    count  = 0
    for cell in df[col]:
        if cell == '' or ' ' in cell:
            df = df.drop([count])       
        else:
            df.at[count, col] = float(cell)
        count += 1

    return df.convert_dtypes()


ltdb = pd.DataFrame({'tractID': data3['trtid10'],
                        'population': data3['l_total'],
                        'non-hisp white': data3['l_nhw'],
                        'median household income': data3['l_medhhinc'],
                        'occupied housing': data3['l_occunit'],
                        'home owners': data3['l_ownerocc'],
                        'renters': data3['l_renterocc'],
                        '% of home owners': data3['l_pctowner'],
                        '% of non-hisp whites': data3['l_pctnhw']})
# display(ltdb.head(10))

ltdb = dtype_conversion(ltdb, 'median household income')

dp = pd.DataFrame({'tractID': data3['trtid10'],
                        'population': data3['total'],
                        'non-hisp white': data3['nhw'],
                        'median household income': data3['medhhinc'],
                        'occupied housing': data3['occunit'],
                        'home owners': data3['ownerocc'],
                        'renters': data3['renterocc'],
                        '% of home owners': data3['pctowner'],
                        '% of non-hisp whites': data3['pctnhw']})

        
dp = dtype_conversion(dp, 'median household income')
# display(dp.drop_duplicates(subset=['tractID']))

m_t, m_p = [], []
total = 0
for i in range(len(dp)):
    m_t.append((dp['population'][i]-dp['non-hisp white'][i]))
    
    
    
    
display(dp)
```
