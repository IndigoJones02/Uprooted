* **pctowner**:% of home owners from LTDB-DP


```python
path3 = "/Users/hbcuc2-miles/Desktop/Gen-Datasets/LTDB-Stats/LTDB_DP.xlsx"
data3 = pd.read_excel(path3)
# display(data3)

def dtype_conversion(df, col):
    df[col] = df[col].astype(str)
    
    count  = 0
    for cell in df[col]:
        if cell == '' or ' ' in cell:
            df = df.drop([count])       
        else:
            df.at[count, col] = float(cell)
        count += 1

    return df.convert_dtypes()


ltdb = pd.DataFrame({'tractID': data3['trtid10'],
                        'tract_population': data3['l_total'],
                        'non-hisp white': data3['l_nhw'],
                        'median household income': data3['l_medhhinc'],
                        'occupied housing': data3['l_occunit'],
                        'home owners': data3['l_ownerocc'],
                        'renters': data3['l_renterocc'],
                        '% of home owners': data3['l_pctowner'],
                        '% of non-hisp whites': data3['l_pctnhw']})
# display(ltdb.head(10))

ltdb = dtype_conversion(ltdb, 'median household income')

dp = pd.DataFrame({'tractID': data3['trtid10'],
                        'tract_population': data3['total'],
                        'non-hisp white': data3['nhw'],
                        'median household income': data3['medhhinc'],
                        'occupied housing': data3['occunit'],
                        'home owners': data3['ownerocc'],
                        'renters': data3['renterocc'],
                        '% of home owners': data3['pctowner'],
                        '% of non-hisp whites': data3['pctnhw']})

        
dp = dtype_conversion(dp, 'median household income')
print(dp.dtypes)
```
